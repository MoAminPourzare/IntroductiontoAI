```python
column = DataFrame.select_dtypes(include="number").columns.drop("Survived")
DataFrame[column] = (DataFrame[column] - DataFrame[column].mean()) / DataFrame[column].std()
display(DataFrame.drop(['Name', 'PassengerId', 'Ticket'], axis=1))
```

```python
print("For sex :")
srv=DataFrame[(DataFrame['Survived'] == 1)].select_dtypes(include = 'number')
die=DataFrame[(DataFrame['Survived'] == 0)].select_dtypes(include = 'number')
x = np.arange(-10, 10, 0.0001)
plt.xlabel('Sex')    
plt.ylabel('Survived')
srvPdf = norm.pdf(x,srv['Sex'].mean(),srv['Sex'].std())
diePdf = norm.pdf(x,die['Sex'].mean(),die['Sex'].std())
plt.plot(x, srvPdf,color = 'gold')
plt.plot(x, diePdf,color = 'green')
plt.grid(True, alpha=0.3, linestyle="--")
plt.show()
```
